```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from jupyter_datatables import init_datatables_mode
init_datatables_mode()
```

```python
accident = pd.read_csv("aviation-accident.csv")
```

```python
accident
```

```python
accident.info()
```

```python
accident.describe()
```

```python
accident = accident.drop(accident[accident['year']=='unknown'].index)
accident = accident.drop(accident[accident['location']=='unknown'].index)
accident = accident.drop(accident[accident['fatalities']=='unknown'].index)
accident = accident.drop(accident[accident['date']=='unknown'].index)
accident = accident.drop(accident[accident['registration']=='unknown'].index)
```

```python
accident.shape
```

```python
accident.dropna(inplace=True)
```

```python
null_counts = accident.isnull().sum()
null_counts
```

```python
location_accidents = accident.groupby('location').size()
fig= px.histogram(location_accidents,x=location_accidents.index,y=location_accidents.values,title='Location Vs Accident')
fig.show()
```

```python
year = accident.groupby('year').size()
fig= px.histogram(year,x=year.index,y=year.values,color=year.index,
            labels = {
                'index' : 'year',
                'value' : 'Accident'
            },title= 'Year Vs Accident')
fig.show()
```

```python
top_countries = accident['country'].value_counts().nlargest(20)
fig= px.histogram(top_countries,x=top_countries.index,y=top_countries.values,
                labels={
                     "index": "Top_Countries",
                 },color=top_countries.index,title='Top Countries Accident')
fig.show()
```

```python
fig = px.histogram(accident,x='country',y='fatalities',
            labels = {
                'counry' : 'Countries',
            },title='Country Vs Fatalities (1922-2023)')
fig.show()
```

```python
accident['date'] = pd.to_datetime(accident['date'], errors='coerce')
yearly_accidents = accident.groupby(accident['date'].dt.year)['cat'].count()

fig = px.scatter(yearly_accidents,x=yearly_accidents.index,y=yearly_accidents.values,color=yearly_accidents.values,
            labels = {
                'date' : 'Year',
                'y' : 'Accident'
            },title='Year Vs Accident Count (1922-2023)')
fig.show()
```

```python
top_aircraft = accident['type'].value_counts().nlargest(20)
fig = px.bar(top_aircraft,color=top_aircraft.index,color_discrete_sequence=['brown','red','green','blue','tan','tomato'],
            labels = {
                'index' : 'Aircrafts',
                'value' : 'Accident'
            },title='Aircrafts Vs Accident (1922-2023)')
fig.show()
```
